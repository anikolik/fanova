```python
import os
import numpy as np
import pandas as pd
```

```python
algorithm_name = "modCMA-ES"  # modDE
if algorithm_name == "modCMA-ES":
    n_configurations = 324
else: 
    n_configurations = 576
dimensions = [5, 30]
target = "unlog"
budgets = {dim: [50*dim, 100*dim, 300*dim, 500*dim, 1000*dim, 1500*dim] for dim in dimensions}
budgets 
```

```python
directory = f"Data/{algorithm_name}/log1"
os.makedirs(directory, exist_ok=True)
```

```python
for dim in dimensions:
    print(dim)
    for budget in budgets[dim]: 
        for configuration_id in np.arange(n_configurations):
            df = pd.read_csv(f"Data/{algorithm_name}/{target}/budget_{budget}_conf_{configuration_id}_{dim}D.csv", index_col=["Unnamed: 0"])
            df["target"] = np.log10(df["target"] + 1)
            print(df)
            df.reset_index().to_csv(f"{directory}/budget_{budget}_conf_{configuration_id}_{dim}D.csv", index=False)
```
